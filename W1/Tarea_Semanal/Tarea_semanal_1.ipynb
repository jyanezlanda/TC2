## 2. ¿Que tipo de filtro es?
Por el diagrama de polos y ceros podemos rapidamente darnos cuenta de que se trata de un filtro pasa todo que posee una ganancia (o atenuación dependiendo de los valores de los componentes) de R2/R1 en bajas frecuencias, sin embargo a altas frecuencias este efecto se vuelve despreciable y la ganancia es unitaria (0dB).

## 3. Norma de frecuencia
Para la norma de frecuencia se utilizó como parámetro 1/C1R3 ya que este coeficiente era el polo del sistema y estaba presente en el numerador, permitiendo asi normalizar el circuito. Por otro lado ciruitalmente representa la constante temporal de la rama C1R3.


## 4. Simulación de la transferencia normalizada

```python
# Librerías externas NumPy, SciPy y Matplotlib
from scipy.signal import TransferFunction
import matplotlib.pyplot as plt
import numpy as np
from pytc2.sistemas_lineales import pzmap, GroupDelay, bodePlot
```

```python
R1 = 10000
R2 = [2*R1, R1, R1/2]
for i in range(len(R2)):
    my_tf = TransferFunction( [1,-R2[i]/R1], [1, 1] )
    bodePlot(my_tf, fig_id=1)
    pzmap(my_tf, fig_id=2) #S plane pole/zero plot
```

## 5. Simulación de la transferencia sin normalizar
A continuación se simulará la transeferencia con los siguientes valores de componentes:
$R1=R=5k \Omega, R3=1k \Omega$  y  $C3=1\mu F$

```python
R1 = 5000
R2 = R1
R3 = 1000
C1 = 1*10**-6

my_tf = TransferFunction( [1,-R2/(R1*C1*R3)], [1, 1/(C1*R3)] )

bodePlot(my_tf, fig_id=1)

pzmap(my_tf, fig_id=2) #S plane pole/zero plot
```
