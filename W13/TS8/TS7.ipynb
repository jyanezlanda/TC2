### Filtro IIR

```python
# Inicialización e importación de módulos

# Módulos para Jupyter
import warnings
warnings.filterwarnings('ignore')

# Módulos importantantes
import numpy as np
import scipy.signal as sig
import matplotlib as mpl
import matplotlib.pyplot as plt
from pytc2.sistemas_lineales import plot_plantilla
import scipy.io as sio


###
## Señal de ECG registrada a 1 kHz, con contaminación de diversos orígenes.
###

# para listar las variables que hay en el archivo
mat_struct = sio.loadmat('ecg.mat')

ecg_one_lead = mat_struct['ecg_lead']
ecg_one_lead = ecg_one_lead.flatten()
cant_muestras = len(ecg_one_lead)

fig_sz_x = 10
fig_sz_y = 7
fig_dpi = 100 # dpi

fig_font_size = 16
```

```python
fs = 1000 # Hz
nyq_frec = fs / 2

ws1 = 1 #Hz
wp1 = 3 #Hz
wp2 = 25 #Hz
ws2 = 35 #Hz

fs = 1000 #kHz
nyq_frec = fs / 2

Amin = 40 #dB
Amax = 0.4 #dB

frecs = np.array([0.0,         ws1,         wp1,     wp2,     ws2,         nyq_frec   ]) / nyq_frec
gains = np.array([-Amax, -Amax, -Amin, -Amin, -Amax, -Amax])
print("Frecuencias claves de la plantilla: ", frecs)
print("Atenuación en dB para cada frecuencia:", gains)

gains = 10**(gains/20)
```

```python
### DISEÑO DEL FILTRO IIR

wpass = [wp1/fs, wp2/fs]
wstop = [ws1/fs, ws2/fs]

sos_pb_dig = sig.iirdesign(wp= wpass, ws= wstop, gpass= Amax, gstop= Amin, analog= False, ftype= 'butter', output='SOS', fs= 1)

ww, hh = sig.sosfreqz(sos= sos_pb_dig, worN= 2000, whole= False, fs = fs)

print("Orden del filtro: ", sos_pb_dig.shape)
```

El filtro cuenta con 15 secciones de segundo orden, por lo que cuenta con 30 polos dentro de la circunferencia unitaria, lo que implica que su orden sea de N = 30.

```python
### GRÁFICO DE LA RESPUESTA DE MÓDULO DEL FILTRO DISEÑADO

plt.figure(1)
plt.title("Filtro IIR: Módulo")
plt.ylabel('Módulo [dB]')
plt.xlabel('Frecuencia [Hz]')
plt.grid(visible=True)
plt.plot(ww, 20*np.log(np.abs(hh)));
plt.axis([0, 100, -100, 5 ])

plot_plantilla(filter_type = 'bandpass', fpass = [wp1, wp2], ripple = Amax , fstop = [ws1, ws2], attenuation= Amin, fs = 1000)
```

```python
plt.figure(2)
plt.title("Filtro IIR: Fase")
plt.ylabel('Fase [rad]')
plt.xlabel('Frecuencia [Hz]')
plt.grid(visible=True)
plt.plot(ww, np.angle(hh));
plt.axis([0, 100, -3.5, 3.5 ])
```

```python
b,a = sig.sos2tf(sos_pb_dig)
w, gd = sig.group_delay((b, a), w=1000, whole=False, fs=fs)

plt.figure(3)

plt.title('Filtro IIR: Retardo de grupo')
plt.plot(w, gd)
plt.ylabel('Retardo de grupo [muestras]')
plt.xlabel('Frecuencia [rad/muestras]')
plt.grid()
plt.axis([0, 200, -15, 5 ])
```

```python
#### Respuesta al impulso del filtro IIR

imp = sig.unit_impulse(2000)
response = sig.sosfilt(sos_pb_dig, imp)

plt.plot(np.linspace(0,1, 2000), response*10)
plt.plot(np.linspace(0,1, 2000), imp)

print("Como la respuesta es pequeña en amplitud se amplifica poor 10 para que sea mas apreciable")

plt.xlabel('Numero de muestra [muestras]')
plt.ylabel('Amplitud')
plt.title('Filtro IIR: Respuesta al impulso')

plt.grid(True)
plt.show()
```

Con estos gráficos queda concluido el diseño del filtro IIR, se pudo lograr que este entre dentro de los límites de la plantilla. Como es de esperarse en estos tipos de filtros los gráficos de fase y retardo de grupo pueden generar distorsión, esto se debe principalmente a que los ceros de estos filtros se ubican en cualquier lugar dentro de la circunferencia unitaria generando estos tipos de gráficas de fase y retardo. Se observará como en el caso de los filtros FIR los gráficos de fase y retardo serán mucho mas amenos ya que sus polos se ubicarán en el origen unicamente.


## Filtro FIR

Para el caso del filtro FIR se utilizará el método de ventana, este consta de multiplicar la respuesta en muestras del brick wall con una ventana que tendrá una forma específica para generar que el h[n] sea finito, además se genera un desplazamiento para tener un filtro causal. En el plano transformado, esta multiplicación por una ventana implica que se convolusionen los espectros del brick wall con el de la ventana elegida, provocando que el brick wall deje de ser ideal.

Según un estudio la ventana que [mejor performance](https://ieeexplore.ieee.org/abstract/document/7322780/) tiene para una señal de ECG es la kaiser, esta ventana tiene la siguiente forma temporal:

<center>  $W(n) = \frac{I_0\left(\beta \sqrt{1 - \left(\frac{2n}{N-1} - 1\right)^2}\right)}{I_0(\beta)}$ </center> 

A la hora de su implementación en python la librería Scipy pide que se incluya el parámetro $\beta$, este parámetro afecta la forma y las características de la ventana. La ventana de Kaiser es una ventana diseñada para proporcionar un buen compromiso entre la anchura del lóbulo principal y la atenuación de los lóbulos laterales en el diseño de filtros FIR. Se fue modificando este parámetro de forma iterativa para poder llegar al mejor resultado.

A continuación se muestra la obtención del filtro en python. Otra vez se utilizó un proceso iterativo para poder encontrar el mejor resultado posible.

```python
### DISEÑO DEL FILTRO FIR
frecs = np.array([0.0,         ws1,         wp1,     wp2,     ws2,         nyq_frec   ]) / nyq_frec
gains = np.array([-Amin, -Amin, -Amax, -Amax, -Amin, -Amin])
gains = 10**(gains/20)

cant_coeficientes = 3501 #Orden del filtro

beta = 5.85
num_win = sig.firwin2(cant_coeficientes, frecs, gains , window= ('kaiser', beta))

den = 1.0

# muestreo el filtro donde me interesa verlo según la plantilla.
w  = np.append(np.logspace(-1, 0.8, 250), np.logspace(0.9, 1.6, 250) )
w  = np.append(w, np.linspace(110, nyq_frec, 100, endpoint=True) ) / nyq_frec * np.pi

_, hh_win = sig.freqz(num_win, den, w)

# renormalizo el eje de frecuencia
w = w / np.pi * nyq_frec

plt.plot(w, 20 * np.log10(abs(hh_win)), label='FIR-Win {:f}'.format(beta))

plt.title('Filtros FIR: Módulo')
plt.xlabel('Frecuencia [Hz]')
plt.ylabel('Módulo [dB]')
plt.grid()
plt.axis([0, 100, -60, 5 ]);

axes_hdl = plt.gca()
axes_hdl.legend()

plot_plantilla(filter_type = 'bandpass', fpass = frecs[[2, 3]]* nyq_frec, ripple = Amax , fstop = frecs[ [1, 4] ]* nyq_frec, attenuation = Amin, fs = fs)
```

```python
plt.figure(2)
phase=np.angle(hh_win)
plt.plot((w / np.pi * nyq_frec), phase)

plt.title('Filtro FIR: Fase')
plt.xlabel('Frecuencia [Hz]')
plt.ylabel('Fase [rad]')
plt.grid()
plt.axis([0, nyq_frec, -2*np.pi, 2*np.pi ])
```

```python
w, gd = sig.group_delay((num_win, den), w=1000, whole=False, fs=fs)

plt.title('Filtro FIR: Retardo de grupo')
plt.plot(w, gd)
plt.ylabel('Retardo de grupo [muestras]')
plt.xlabel('Frecuencia [rad/muestras]')
plt.grid()
plt.axis([0, nyq_frec, 0, 2000])
```

```python
imp = sig.unit_impulse(3000)

den=np.zeros(5001)
den[0]=1

respuesta = sig.lfilter(num_win, den, imp)

print("Como la respuesta es pequeña en amplitud se amplifica poor 10 para que sea mas apreciable")

plt.plot(np.linspace(0,1, 3000), respuesta*(10))
plt.plot(np.linspace(0,1, 3000), imp)

plt.xlabel('Numero de muestra [muestras]')
plt.ylabel('Amplitud')
plt.title('Filtro FIR: Respuesta al impulso')

plt.grid(True)
plt.show()
```

El filtro no termina de entrar por completo en la plantilla, esta tiene requisitos altos, se podría haber continuado subiendo el orden del filtro y jugando con el parámetro $\beta$ sin embargo el error es bastante pequeño por lo que se decidió dejarlo así. Es destacable como era de esperarse que los gráficos de fase y retardo son muy buenos en este tipo de filtros, cabe destacar que se eligió un orden impar del filtro para que el retardo quede en un número entero de muestras, recordando que el retardo de los filtros FIR siempre será $\frac{N-1}{2}$ siendo N el orden del filtro. Como era esperable la fase es lineal.

Finalmente implementaremos **los mismos filtros** con una técnica diferente llamada *filtrado bidireccional* *[scipy.signal.filtfilt](https://docs.scipy.org/doc/scipy-0.18.1/reference/generated/scipy.signal.filtfilt.html)*.

El filtrado bidireccional es una técnica sumamente conveniente al momento de filtrar digitalmente una señal, dado que evita por completo cualquier distorsión de fase. Esto afecta principalmente a los filtros IIR, donde las distorsiones pueden ser notorias. Sin embargo, los filtros FIR cuya respuesta al impulso sea simétrica (par o simpar), **no** producen distorsión de fase debido a que su fase **es lineal**. 

En ambos casos, tanto para los IIR como para los FIR, simétricos o no, el resultado es una anulación completa de la respuesta de fase, obteniéndose así un **filtro de fase cero**. Recordemos que la transformada de Fourier de la inversión en tiempo es 

\begin{equation}
x[-k] \stackrel{\mathrm{DFT}}{\longleftrightarrow} X^*(e^{j \, \Omega })
\end{equation}

Por lo tanto, vemos que la inversión en tiempo conlleva la negación de la respuesta de fase de una señal. Entonces utilizaremos una doble inversión temporal para cancelar los efectos de fase, y al mismo tiempo el resultado neto sobre la señal espejada dos veces será nulo

\begin{equation}
Y(e^{j \, \Omega }) = X(e^{j \, \Omega }) \cdot T^*(e^{j \, \Omega }) \cdot T(e^{j \, \Omega }) 
\end{equation}

\begin{equation}
Y = X \cdot |T|^2
\end{equation}

*(Se simplificó la notación para clarificar el resultado)* Se observa como resultado de aplicar el método bidireccional un filtro equivalente al cuadrado del módulo y con aporte nulo de fase para todas las frecuencias. Esto equivale a haber atravesado dos veces por el mismo filtro, por lo tanto la atenuación será el doble que la del filtro original (y el *ripple* en banda también!) razón por lo cual debe modificarse la plantilla de diseño **relajando a la mitad** la atenuación, es decir exigiendo menos atenuación, pero **duplicando** la exigencia de ripple, es decir dividiendo por dos el ripple.

```python
# Procedemos al filtrado

ECG_f_win = sig.filtfilt(num_win, den, ecg_one_lead)

# Segmentos de interés
regs_interes = ( 
        np.array([5, 5.2]) *60*fs, # minutos a muestras
        np.array([12, 12.4]) *60*fs, # minutos a muestras
        np.array([15, 15.2]) *60*fs, # minutos a muestras
        )

for ii in regs_interes:
    
    # intervalo limitado de 0 a cant_muestras
    zoom_region = np.arange(np.max([0, ii[0]]), np.min([cant_muestras, ii[1]]), dtype='uint')
    
    plt.figure(figsize=(fig_sz_x, fig_sz_y), dpi= fig_dpi, facecolor='w', edgecolor='k')
    plt.plot(zoom_region, ecg_one_lead[zoom_region], label='ECG', lw=2)
    #plt.plot(zoom_region, ECG_f_butt[zoom_region], label='Butter')
    plt.plot(zoom_region, ECG_f_win[zoom_region], label='Win')
    
    plt.title('ECG filtering example from ' + str(ii[0]) + ' to ' + str(ii[1]) )
    plt.ylabel('Adimensional')
    plt.xlabel('Muestras (#)')
    
    axes_hdl = plt.gca()
    axes_hdl.legend()
    axes_hdl.set_yticks(())
            
    plt.show()
```

```python
regs_interes = ( 
        [4000, 5500], # muestras
        [10e3, 11e3], # muestras
        )

for ii in regs_interes:
    
    # intervalo limitado de 0 a cant_muestras
    zoom_region = np.arange(np.max([0, ii[0]]), np.min([cant_muestras, ii[1]]), dtype='uint')
    
    plt.figure(figsize=(fig_sz_x, fig_sz_y), dpi= fig_dpi, facecolor='w', edgecolor='k')
    plt.plot(zoom_region, ecg_one_lead[zoom_region], label='ECG', lw=2)
    #plt.plot(zoom_region, ECG_f_butt[zoom_region], label='Butter')
    plt.plot(zoom_region, ECG_f_win[zoom_region], label='Win')
    
    plt.title('ECG filtering example from ' + str(ii[0]) + ' to ' + str(ii[1]) )
    plt.ylabel('Adimensional')
    plt.xlabel('Muestras (#)')
    
    axes_hdl = plt.gca()
    axes_hdl.legend()
    axes_hdl.set_yticks(())
            
    plt.show()
```
