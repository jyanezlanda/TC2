```python
s = sp.symbols('s', complex=True)

S21 = sp.Rational('15')/(s**3 + sp.Rational('6')*s**2 + sp.Rational('15')*s + sp.Rational('15') )

print_latex(a_equal_b_latex_s('S21(s)', S21))

S11sq = sp.factor( 1- (S21 * S21.subs(s, -s)) )
S11sq = sp.simplify(S11sq)
print_latex(a_equal_b_latex_s('S11sq(s)', S11sq))

S11 = modsq2mod_s(S11sq)
print_latex(a_equal_b_latex_s('S11(s)', S11))
```

## Comportamiento del parámetro S11

El S11 es el parámetro que compara la onda de tensión que se refleja en base a la tensión que se inyecta en el puerto de entrada cuando el puerto de salida se encuentra adaptado:

### Centro de la banda de paso:

La banda de paso de este circuito es las bajas frecuencias, en esta banda el S11 tiende a anularse, indicandonós que toda la potencia inyectada se transmite a la carga y poca se refleja al generador.

### Banda de transición y frecuencia de corte:

En la transición y principalmente en la frecuencia de corte el S11 tiene el mismo valor que el S21 lo cual nos indica que la mitad de la potencia se transmite a la carga y la otra mitad se refleja al generador.

### Banda de detenida:

Para altas frecuencias (banda de detinida de la transferecia) el S11 tiende a ser 0dB, esto es un claro indicio de que en alta frecuencia la carga no resive tensión sino que esta se refleja y vuelve al generador.
