## Simulación del ejercicio en clase.

A continuación se deja la simulación de ambos resultados obtenidos al resolver la ecuación para obtener los coeficientes de Cheby.


```python
from scipy.signal import TransferFunction
import matplotlib.pyplot as plt
from pytc2.sistemas_lineales import pzmap, GroupDelay, bodePlot
```

```python
ep = 0.31061

A = 4*ep
B = [1.66 , -1.7]
C = [2.05 , 0.93]

for i in range(len(B)):
    my_tf = TransferFunction ( [1], [A , B[i] , C[i] , 1])
    
    bodePlot(my_tf, fig_id=1, filter_description = 'B={:3.3f}'.format(B[i]) )
    pzmap(my_tf, fig_id=2, filter_description = 'B={:3.3f}'.format(B[i])) #S plane pole/zero plot
    GroupDelay(my_tf, fig_id=3, filter_description = 'B={:3.3f}'.format(B[i]))
```
