## 2) Deducción de la transferencia

Se sabe que se debe construir un filtro notch de 2do orden, gracias a la práctica de todo el cuatrimestre la transferencia de este filtro resulta ser trivial, ya que es conocido que debe contar con ceros sobre el eje $j\omega$ en el valor de frecuencia a eliminar y polos en la misma circunferencia para que el filtro sea simétrico. Luego de este análisis la transferencia obtenida sería la siguiente:


\begin{equation}
\ T_{N}(S) = \frac{s²+\omega_0²}{s²+s\frac{\omega_0}{Q}+\omega_0²} 
\end{equation}

Donde: 

$\omega_0=2\pi*50Hz$

$Q = \frac{f_0}{\Delta f} = \frac{50}{10} = 5$

## 3) Simulación en Python

A continuación se adjunta una simulación en python de la trasnferencia del filtro obtenida, además se adjuntan unas imágenes en detalle de los principales puntos de interes para poder corroborar que esta cumple con los requisitos de la plantilla.

```python
import scipy.signal as sig
import pytc2.sistemas_lineales as tc2
import math as m
```

```python
# Requisitos de plantilla
f0 = 50         # [Hz]
BW = 10         # [Hz]              
n  = 2          # [Hz]

w0 = 2*m.pi*f0
Q = f0/BW

print("Q = ", Q)
```

```python
# Armado de la funcion transferencia normalizada
den = [1, 1/Q, 1]
num = [1, 0, 1]

tf_notch = sig.TransferFunction(num, den)

print("Transferencia del filtro notch normalizado")
tc2.pretty_print_lti(num, den)

# Analisis la transferencia
tc2.analyze_sys(tf_notch, sys_name='Filtro Notch')
```

### 7.3) Mediciones realizadas

En la práctica de laboratorio se realizaron tres mediciones directas con el osciloscopio una vez que se calibraron todas las puntas:

1. **Medición de la tensión RMS de la señal de entrada**
2. **Medición de la tensión RMS de la señal de salida**
3. **Medición del retardo de fase**

Con los primeros dos valores se calcula el módulo de la transferencia simplemente dividiendo el valor de salida por el del entrada, luego a este valor se lo puede logaritmar para obtener la atenuación en dB. Cabe destacar que como se trata de un filtro notch la atenuación en el pozo es muy grande por lo que la señal de salida será muy pequeña y puede tener ruido, por lo que se decidió utilizar la tensión RMS del ciclo para disminuir el ruido en las mediciones. La tercera medición se utilizará para calcular indirectamente el gráfico de fase ya que se puede calcular simplemente $\psi = \Delta t * \omega$. Además con estos valores se realizará el retardo de grupo derivando punto a punto cada medición.
