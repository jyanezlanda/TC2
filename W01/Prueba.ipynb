Tambien se realizó una pequeña simulación en python de la transferencia obtenida se deja a continuación:

```python
# Librerías externas utilizadas
from scipy.signal import TransferFunction
import matplotlib.pyplot as plt
from pytc2.sistemas_lineales import pzmap, GroupDelay, bodePlot
```

```python
R1 = 10000
R2 = R1
R3 = 1000
C1 = 1*10**-6

my_tf = TransferFunction( [1,-R2/(R1*C1*R3)], [1, 1/(C1*R3)] )

plt.close('all')

bodePlot(my_tf, fig_id=1)

pzmap(my_tf, fig_id=2) #S plane pole/zero plot
```

En la simulación observamos que en el mapa de polos y ceros los ejes estan nomenclados al reves y en la gráfica de bode si bien parece que oscila de gran manera cabe destacar que estas osilasciones suceden entre $+-1e-15$ lo cual se puede considerar 0 y la transferencia sería un pasa todo que no atenua ni amplifica, el circuito es práctico ya que si vemos la fase obtenemos una inversión en un rango de frecuencias.
